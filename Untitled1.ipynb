```python
import requests
import pandas as pd

# Step 1: Get data for Jaipur
API_KEY = "470d8152fada130825799f3d4f4a54ca"
CITY = "Jaipur"
url = f"https://api.openweathermap.org/data/2.5/weather?q={CITY}&appid={API_KEY}"

data = requests.get(url).json()

# Step 2: Convert JSON to DataFrame
df = pd.json_normalize(data)
df
```

```python
df_selected= df[[
    "name",
     "dt",
    "coord.lat",
    "coord.lon",
    "main.temp",
    "main.humidity",
    "wind.speed",
    "weather"
]]
df_selected
```

```python
# Convert temperature from Kelvin to Celsius
df_selected["temp_celsius"] = (df_selected["main.temp"] - 273.15).round(1)

# Convert timestamp (dt) into readable date time
df_selected["date_time"] = pd.to_datetime(df_selected["dt"], unit='s')
```

```python
df_selected
```

```python
df_selected["weather_main"]=df_selected["weather"].apply(lambda x:x[0]["main"])
df_selected["description"]=df_selected["weather"].apply(lambda x:x[0]["description"])
```

```python
df_selected
```

```python
def risk(row):
    if row["wind.speed"]>10 or row["temp_celsius"] > 40 or row["temp_celsius"] < 5:
        return "High"
    return "Low"
df_selected["risk_level"]=df_selected.apply(risk,axis=1)
```

```python
df_selected
```

```python
# Drop unwanted technical columns
df_selected_clean = df_selected[[
      "name",
    "dt",
    "coord.lat",
    "coord.lon",
    "main.temp",
    "main.humidity",
    "wind.speed",
    "weather"
]]

df_selected_clean.rename(columns={
    "name": "city",
    "coord.lat": "lat",
    "coord.lon": "lon",
    "main.humidity": "humidity",
    "wind.speed": "wind_mps"
}, inplace=True)

df_selected_clean
```

```python
import os

file_path = "weather_data2.csv"

if os.path.exists(file_path):
    df_selected.to_csv(file_path, mode='a', header=False, index=False)
else:
    df_selected.to_csv(file_path, index=False)
```

```python
cities = ["Delhi", "Mumbai", "Jaipur", "Kolkata", "Bangalore", "Chennai", "Hyderabad", "Pune"]

all_data = []

for city in cities:
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}"
    data = requests.get(url).json()
    df_city = pd.json_normalize(data)   # already normalized
    all_data.append(df_city)

# ✅ Correct way: combine all rows
df_all = pd.concat(all_data, ignore_index=True)

df_all
```

```python
# Select only useful columns from multi-city data
df_clean = df_all[[
    "name",
    "dt",
    "coord.lat",
    "coord.lon",
    "main.temp",
    "main.humidity",
    "wind.speed",
    "weather"
]].copy()

df_clean
```

```python
df_clean["temp_celsius"] = (df_clean["main.temp"] - 273.15).round(1)
df_clean["date_time"] = pd.to_datetime(df_clean["dt"], unit='s')
df_clean

```

```python
df_clean["weather_main"] = df_clean["weather"].apply(lambda x: x[0]["main"])
df_clean["description"] = df_clean["weather"].apply(lambda x: x[0]["description"])
df_clean
```

```python
def risk(row):
    if row["temp_celsius"] > 40 or row["wind.speed"] > 10:
        return "High"
    return "Low"

df_clean["risk_level"] = df_clean.apply(risk, axis=1)
df_clean
```

```python
df_final = df_clean[[
    "name",
    "dt",
    "coord.lat",
    "coord.lon",
    "main.temp",
    "main.humidity",
    "wind.speed",
    "weather",
    "temp_celsius",
    "date_time",
    "weather_main",
    "description",
    "risk_level"
]]

df_final.rename(columns={
    "name": "city",
    "coord.lat": "lat",
    "coord.lon": "lon",
    "main.humidity": "humidity",
    "wind.speed": "wind_mps"
}, inplace=True)

df_final


```
